```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
games = pd.read_csv('vgsales.csv', sep=',')
```

```python
games.head()
```

```python
games.groupby('Genre').mean()
```

Sales are difficult to calculate corralations with, because the numbers are difficult to predict. You can tell which ones are more populaire, but not corrolations between numbers/genres. You can just read easy which region does better than the other.
