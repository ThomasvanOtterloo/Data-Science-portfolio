```python
import pandas as pd
import seaborn as sns
games = pd.read_csv('vgsales.csv', sep=',')
```

```python
games.head()
```

```python
games = games.corr()
games.style.background_gradient(cmap='coolwarm', axis=None).format(precision=2)
```

```python
ax = games.drop(['Name', 'Platform', 'Rank', 'Year', 'JP_Sales', 'Other_Sales','Global_Sales'], axis=1).plot(figsize=(10,10))
ax.set_xlim(left=0, right=150)
```

```python
games.plot(kind='scatter', x='NA_Sales', y='EU_Sales')
```

This matched my expectations because both regions are very similar big on the game industry. The sales are very similar.
